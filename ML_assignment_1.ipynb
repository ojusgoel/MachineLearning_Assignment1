For the Wine Quality Data Set, we have to convert all the values in the quality attribute to
0 (bad) if the value is less than or equal to 6 and to 1 (good) otherwise and the normalize
all the other attributes between 0 and 1 by min-max scaling.

```python
# Standard libraries import
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
# Reading the Wine Quality Data Set
df = pd.read_csv('WineQT.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
count_greater_than_6 = (df['quality'] > 6).sum()
print(count_greater_than_6)
```

```python
# Changing the value to 0 if the value is less than or equal to 6 and 0 otherwise

df['quality'] = df['quality'].apply(lambda x: 0 if x <= 6 else 1)
df
```

```python
count_greater_than_0 = (df['quality'] > 0).sum()
print(count_greater_than_0)
```

```python
# Instantiating MinMaxScaler
scaler = MinMaxScaler()

# Normalising the data
scaled_df = scaler.fit_transform(df.drop('quality', axis=1))

# Normalized data
scaled_df
```

```python
# Converting back to a Data Frame

normalised_df = pd.DataFrame(scaled_df, columns=df.drop('quality', axis=1).columns)
normalised_df['quality'] = df['quality']
normalised_df.head()
```

```python
# Exporting the Normalise DF to a .csv file

normalised_df.to_csv('WineQT_Normalised.csv', index=False)
```
